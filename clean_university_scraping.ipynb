```python
import pandas as pd
df=pd.read_csv('location_5.22.csv')
df
```

```python
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
df
```

```python
locator = Nominatim(user_agent="myGeocoder", timeout=10)
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
```

```python
from tqdm.notebook import tqdm 

tqdm.pandas()

df['full location']=df['Location'].progress_apply(geocode)
df
```

```python
df[df['full location']=='None']
```

```python
df['latitude']=df['full location'].apply(lambda loc: loc.latitude if loc else None)
df['longitude']=df['full location'].apply(lambda loc: loc.longitude if loc else None)
df
```

```python
df[df['longitude'].isna()]
```

```python
import geopandas as gpd

# Assuming december_df is your DataFrame with 'Longitude' and 'Latitude' columns
geometry = gpd.points_from_xy(df.longitude, df.latitude)
gdfa = gpd.GeoDataFrame(df, crs='EPSG:4326', geometry=geometry)

# Display the GeoDataFrame
gdfa
```

```python
import folium
import geopandas as gpd

# Remove rows with NaN latitude or longitude
gdfa = gdfa.dropna(subset=['latitude', 'longitude'])

# Create the map
m = folium.Map(location=[20, 0], zoom_start=2, width='800px', height='500px')

# Add markers with smaller icons
for idx, row in gdfa.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"""
        <b>University Name:</b> {row['University Name']}
        <br>
        <b>Financial Package for International Students:</b> {row['Financial Package for International Students']}
        <br>
        <b>Total Cost:</b> {row['Total Cost']}
        """,
        icon=folium.Icon(icon='university', prefix='fa', icon_size=(25, 25))  # Smaller icon size
    ).add_to(m)


# Display the map
m
```
